```
import pandas as pd 
import xml.etree.ElementTree as etree
```

```
data = pd.read_csv("/content/drive/My Drive/mart_export.csv")
```

```
data.head()
```

```
def geneexpansion(x):
  S=[]
  flag=0
  for t in x:
    y=" "
    r=[]
    for index, row in data.iterrows():
      if(str(row["Gene name"])== t or str(row['Gene Synonym'])==t):
        m=str(row["Gene name"])+" "+str(row["Gene Synonym"])
        flag=1
      if(flag==1):
        y+=" "+m
    if(flag==1):
      r=list(set(list(y.split(" "))))
      S.append(" ".join(r))
      flag=0
      continue
    if(flag==0):
      S.append(t)
  return(" ".join(S))
   
```

```
tree = etree.parse("/content/drive/My Drive/topics2018.xml")
root = tree.getroot()
```

```
columns = ["num","disease", "gene","demographic"]
df = pd.DataFrame(columns = columns)
```

```
count=0
for node in root: 
  num=node.find("num").text
  disease = node.find("disease").text 
  gene = node.find("gene").text
  demographic = node.find("demographic").text 
  df = df.append(pd.Series([num, disease, gene, demographic], index = columns), ignore_index = True)

```

```
len(df)
```

```
T=[]
for ind in df.index: 
     S=""
     l=df['gene'][ind]
     x=l.split(" ")
     S=geneexpansion(x)
     print(x)
     T.append(S)
```

```
df['Gene_Expansion']=T
```
